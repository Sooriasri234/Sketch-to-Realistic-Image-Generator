```python
import torch
print("CUDA Available:", torch.cuda.is_available())
print("GPU Name:", torch.cuda.get_device_name(0))
```

```python
from diffusers import StableDiffusionControlNetPipeline, ControlNetModel
import torch

controlnet = ControlNetModel.from_pretrained(
    "lllyasviel/sd-controlnet-scribble", torch_dtype=torch.float16
)

pipe = StableDiffusionControlNetPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    controlnet=controlnet,
    torch_dtype=torch.float16
)
pipe.to("cuda")
```

```python
import gradio as gr

def sketch_to_real(prompt, sketch_img):
    result = pipe(prompt, image=sketch_img, num_inference_steps=30).images[0]
    return result

with gr.Blocks() as demo:
    gr.Markdown("🎨 **Sketch to Realistic Jewelry Generator**")

    with gr.Row():
        with gr.Column():
            prompt_input = gr.Textbox(label="Enter Prompt")
            sketch_input = gr.Image(label="Upload Sketch", type="pil")
            submit_btn = gr.Button("Generate")

        with gr.Column():
            output_img = gr.Image(label="Generated Realistic Image")

    submit_btn.click(fn=sketch_to_real, inputs=[prompt_input, sketch_input], outputs=[output_img])

demo.launch(share=True)
```
